## Simulating a random walk

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.stattools import acf

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Generate a dataset of 1000 data points
- The starting point is 0, and we add random noise to the previous point to generate the next one

```python
steps = np.random.standard_normal(1000)
steps[0] = 0
random_walk = np.cumsum(steps)
```

### Plotting our dataset, we see the following

```python
plt.figure(figsize=(10, 7.5))
plt.plot(random_walk)
plt.title("Simulated Random Walk")
plt.show()
```

### Plot autocorrelation plot (correlogram)

```python
random_walk_acf_coef = acf(random_walk)
plot_acf(random_walk, lags=30)
```

### Take difference between each point to get a purely random stochastic process

```python
random_walk_diff = np.diff(random_walk, n=1)
```

```python
plt.figure(figsize=(10, 7.5))
plt.plot(random_walk_diff)
plt.title("Noise")
plt.show()
```

```python
plot_acf(random_walk_diff, lags=30)
```

## Moving Average Process

```python
ar2 = np.array([2])
```

```python
ma2 = np.array([1, 0.9, 0.3])
```

```python
ma2_process = ArmaProcess(ar2, ma2).generate_sample(nsample=1000)
```

```python
plt.figure(figsize=(10, 7.5))
plt.plot(ma2_process)
plt.title("Moving Average Process of Order 2")
plt.show()
```
